Q.2b

```python
ds = xr.open_dataset("/mnt/Metservice_WRF/RH_20200101000000.nc") #Reading the relative humdity in the netcdf file
```

```python
time_slice_rh=ds.RH.sel(time=slice("2020-01-01","2020-01-31")) #Relative humidity time slice for the month of January in 2020
```

```python
time_slice_rh.sel(lat=45.88,lon=170.5,method="nearest").plot(linestyle="solid", color="orange",figsize=(12,5))
plt.xlabel("Time",fontsize=20)
plt.ylabel("RH (%)",fontsize=20)
plt.xticks(fontsize=14,rotation=45)
plt.yticks(fontsize=14,rotation=0)
plt.title("Dunedin Relative Humidity Time Series for the Month of January 2020") 
```

Q.2c

```python
time_slice_t=ts.T.sel(time=slice("2020-01-01","2020-01-31")) #Temperature time slice for the month of January in 2020
```

```python
time_slice_t.sel(lat=45.88,lon=170.5,height=20,method="nearest").plot(linestyle="solid", color="orange",figsize=(12,5))
plt.xlabel("Time",fontsize=20)
plt.ylabel("T (K)",fontsize=20)
plt.xticks(fontsize=14,rotation=45)
plt.yticks(fontsize=14,rotation=0)
plt.title("Dunedin Temperature Time Series at 20m for the Month of January 2020")
```
